# Data Map and Scaling

```python
ds.groupby(by=['purpose']).count()
```

```python
ds['_history'] = ds['history'].apply({'terrible':0, 'poor':1,'good':2}.get).astype(float)
ds['_sex'] = ds['sex'].apply({'male':0, 'female':1}.get).astype(float)
ds['_employ'] = ds['employ'].apply({'0-1year':0, '1-7years':1,'7+years':2,'unemployed':3}.get).astype(float)
ds['_housing'] = ds['housing'].apply({'forfree':0, 'own':1,'rent':2}.get).astype(float)
ds['_foreign'] = ds['foreign'].apply({'foreign':0, 'german':1}.get).astype(float)
ds['_purpose'] = ds['purpose'].apply({'biz':0, 'edu':1,'goods/repair':2,'newcar':3,'usedcar':4}.get).astype(float)


#Scaler

amount_scaler=StandardScaler()
ds['_amount']=amount_scaler.fit_transform(ds[['amount']])
duration_scaler=StandardScaler()
ds['_duration']=amount_scaler.fit_transform(ds[['duration']])
age_scaler=StandardScaler()
ds['_age']=amount_scaler.fit_transform(ds[['age']])
```

# Apply K-means

```python
X = ds[['Default','_amount','_duration','installment','cards','liable','_history','_sex','_employ','_housing','_foreign','_purpose','_age']]
```

```python
#X = ds[['Default','_amount', '_duration']]
```

Escoji 3 clusters para agrupar los datos en 3 potenciales categorias:
1-Aquellos que van a pagar
2-Aquellos que no van a pagar(default)
3-Talvez,puede ser que si paguen o no

```python
kmeans = KMeans(n_clusters=3, random_state=12).fit(X)
```

```python
#centroids = kmeans.cluster_centers_
centroids = pd.DataFrame( kmeans.cluster_centers_, columns=X.columns)
```

Aplicando PCA

```python
X_train=X
```

```python
X = X_train.drop(['Default'], axis=1)
y = X_train['Default']

pca = PCA(n_components=12)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
credit_pca = pd.DataFrame(
    data    = pca.components_,
    columns = X.columns,
    index   = ['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12']
)
credit_pca
```

```python
fig = px.imshow(credit_pca)
fig.show()
```

```python
pca.components_
```

```python
X_pca
```

```python
fig = px.scatter(x=X_pca[:, 0], y=X_pca[:, 1], color=y)
fig.show()
```

```python
fig = px.scatter_3d(x=X_pca[:, 0], y=X_pca[:, 1], z=X_pca[:, 2], color=y)
fig.show()
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
fig = go.Figure()

var_acum = pca.explained_variance_ratio_.cumsum()

fig.add_trace(go.Scatter(
    x=[x for x in range(len(var_acum))], 
    y=var_acum,
    mode='lines+markers'
))

fig.update_layout(
    title="Porcentaje de varianza explicada acumulada",
    xaxis_title="Componentes Principales",
    yaxis_title="% Varianza acumulada"
)

fig.show()
```
